# Olive Leaf Disease Classification using MobileNetV2

## 1. Introduction

This notebook demonstrates the training and evaluation of a convolutional neural network for detecting diseases in olive leaves using transfer learning with MobileNetV2.

- Classes: aculus_olearius, healthy, peacock_spot
- Goal: Classify images into one of the above categories accurately and robustly.

```python
# Imports
import os
import sys
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.metrics import classification_report, confusion_matrix
from IPython.display import Image, display

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
```

## 2. Data Preprocessing & Augmentation

Data is loaded using Keras' ImageDataGenerator with heavy augmentations applied to improve generalization and balance the classes.

- Rescale: 1./255
- Augmentations: rotation, zoom, brightness, flip, shear, etc.
- Validation split: 20%

```python
# Display example images from the training generator
from src.data_loader import get_data_generators

train_gen, val_gen, test_gen = get_data_generators("../data", image_size=(224, 224), batch_size=32)

class_names = list(train_gen.class_indices.keys())

plt.figure(figsize=(12, 6))
for i in range(6):
    batch = next(train_gen)
    image = batch[0][0]
    label = class_names[np.argmax(batch[1][0])]
    plt.subplot(2, 3, i+1)
    plt.imshow(image)
    plt.title(f"Label: {label}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 3. Model Architecture

The model is based on MobileNetV2 pretrained on ImageNet, with a custom classification head:

- Base: MobileNetV2 (frozen BatchNorm layers)  GlobalAveragePooling2D  Dropout
- Dense(256)  LeakyReLU  BatchNorm  Dropout
- Dense(128)  LeakyReLU
- Output: softmax layer for 3 classes

Loss: categorical_crossentropy  
Optimizer: Adam  
Metrics: accuracy, ROC-AUC, precision, recall, top-2 accuracy

```python
from src.model_builder import build_model

model = build_model(input_shape=(224, 224, 3), num_classes=3)
model.summary()
```

## 4. Training

Training uses:

- Class weights to handle imbalance
- EarlyStopping to avoid overfitting
- ReduceLROnPlateau to reduce learning rate on plateaus
- ModelCheckpoint to save the best model

30 epochs max.

```python
# Load history
with open("../outputs/history/history.pkl", "rb") as f:
    history = pickle.load(f)

# Plot accuracy and loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history["accuracy"], label="Train Acc")
plt.plot(history["val_accuracy"], label="Val Acc")
plt.title("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history["loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")
plt.title("Loss")
plt.legend()
plt.tight_layout()
plt.show()
```

## 5. Evaluation on Test Set

We evaluate the model using:

- Confusion Matrix
- Classification Report
- ROC-AUC score
- PR Curve
- Misclassified samples

```python
# Load model
model = load_model("../outputs/models/olive_leaf_disease_model.h5", compile=False)

# Predict
y_true = test_gen.classes
y_pred_probs = model.predict(test_gen, verbose=0)
y_pred = np.argmax(y_pred_probs, axis=1)
```

```python
# Classification Report
print(classification_report(y_true, y_pred, target_names=class_names))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```
